```r
library(tidyverse)
```

```r
df=read.csv("Salary_Data.csv")
df
```

```r
install.packages("caTools")
```

```r
library(caTools)
```

```r
set.seed(321)

```

```r
split=sample.split(df$Salary,SplitRatio =0.8)
```

```r
training=subset(df,split==T)
```

```r
test=subset(df,split==F)
```

```r
print(training)
```

Creating regressor 

```r
regressor=lm(formula=Salary~ YearsExperience,data=training)
```

```r
regressor
```

```r
#salary=24416+9587x years of experience
```

```r
y_pred=predict(regressor,newdata = test)
print(y_pred)
```

```r
# Visualising the Training set results
library(ggplot2)
```

```r
ggplot()+
  geom_point(aes(x=training$YearsExperience,y=training$Salary))+
  geom_line(aes(x=training$YearsExperience,y=predict(regressor,newdata=training)),col="blue")+
  ggtitle("Salary Vs Experience on Training Set")+
  xlab("Experience")+
  ylab("Salary")
```

```r
ggplot()+
  geom_point(aes(x=test$YearsExperience,y=test$Salary))+
  geom_line(aes(x=training$YearsExperience,y=predict(regressor,newdata=training)),col="blue")+
  ggtitle("Salary Vs Experience on test Set")+
  xlab("Experience")+
  ylab("Salary")
```
