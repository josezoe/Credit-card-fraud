List of text to store resume and job description

```python
text=[resume,job_desc]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
count_matrix=cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(count_matrix))
```

```python
match=cosine_similarity(count_matrix)[0][1]
match=match*100
match=round(match,2)
print(match)
```

```python
if(match >90):
    print("application is accepted ")
else:
    print("application needs attention")    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# convert documents to vectors
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([resume, job_desc])

# create a dictionary mapping the indices to the words
index_to_word = {i: word for i, word in enumerate(vectorizer.get_feature_names())}

# find the non-similar words
non_similar_words = []
for i in range(vectors[0].shape[1]):
  if vectors[0][0, i] != vectors[1][0, i]:
    word = index_to_word[i]
    non_similar_words.append(word)

print(non_similar_words)
non_similar_words
non_similar_words
```

```python
filtered_words = []
for word in non_similar_words:
  if not any(char.isdigit() for char in word):
    filtered_words.append(word)

non_similar_words = filtered_words
print(non_similar_words)
```

```python
import nltk

# download the list of filler words
nltk.download("stopwords")

# get the list of filler words
filler_words = nltk.corpus.stopwords.words("english")

# remove filler words
filtered_words = [word for word in non_similar_words if word not in filler_words]

non_similar_words = filtered_words
```
